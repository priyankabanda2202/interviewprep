```python
#Correlartion:each of the independent features are collinear o each other that is dependent on each other and also both has so much dpendency onn the output varable

#how to see then whether thay ra colliner or not and what to do in suh cases

#here firstly w are implementing the ordinary least square to find out teeh collinearity among the features
df=pd.read_csv("Advertising.csv")
df2=pd.read_csv("Salary_Data.csv")
X=df[['TV','radio','newspaper']]
Y=df['sales']
```

```python
#fit a OLS  model  with intercept  on TV and Radio
import statsmodels.api as sm
X=sm.add_constant(X)
print(X)
model=sm.OLS(Y,X).fit()
model.summary()
```

```python
#paramters w see here:r2 vlue (0to 1),coef is nothing but y=m1x+m2x+m3x+....here m1,m2,m3 values ar denote by coeffcients,std error should not be a high value,p value should be in range (0,0.05)fo rthe multincolleranity concept 
#hence we say by teh abov datast that none of the features are having teh multicollearnity aong themselves.

import matplotlib.pyplot as plt
X.iloc[:,1:].corr()

#these values should be vry less and hence we say that none of the feates are having the crrelartion among themmselves.all are pretty much nearer to zero.

```
